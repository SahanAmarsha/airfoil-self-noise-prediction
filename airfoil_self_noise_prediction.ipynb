* frequency, angle of attack, and displacement thickness all have considerable amount of outliers since the total dataset record count is 1500. Therefore we cannot remove all outliers.
* Instead we can replace outliers with lower and upper limit values.
* Since this dataset was collected as an experiment by changing the various free stream velocities and chord length, we won't be removing outliers in those variables.
* Let's replace outliers in other independent variables.





```python
# Replacing outliers for Suction Side Displacement Thickness (delta)

fig, axes = plt.subplots(1,2)
plt.tight_layout()

print("Before Shape: ",df.shape)
sns.boxplot(df['suction side displacement thickness (delta)'],orient='v',ax=axes[0])
axes[0].title.set_text("Before")

# Handling outliers
Q1 = df['suction side displacement thickness (delta)'].quantile(0.25)
Q3 = df['suction side displacement thickness (delta)'].quantile(0.75)
IQR = Q3-Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR

df2 = df
df2['suction side displacement thickness (delta)'] = np.where(df2['suction side displacement thickness (delta)']>upper_limit,upper_limit,df2['suction side displacement thickness (delta)'])
df2['suction side displacement thickness (delta)'] = np.where(df2['suction side displacement thickness (delta)']<lower_limit,lower_limit,df2['suction side displacement thickness (delta)'])
print("After:", df2.shape)
sns.boxplot(df2['suction side displacement thickness (delta)'],orient='v',ax=axes[1])
axes[1].title.set_text("After")
plt.show()

df = df2
```

```python
# Replacing outliers for Angle of Attack (o)

fig, axes = plt.subplots(1,2)
plt.tight_layout()

print("Before Shape: ",df.shape)
sns.boxplot(df['angle of attack (o)'],orient='v',ax=axes[0])
axes[0].title.set_text("Before")

# Handling outliers
Q1 = df['angle of attack (o)'].quantile(0.25)
Q3 = df['angle of attack (o)'].quantile(0.75)
IQR = Q3-Q1
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR

df2 = df
df2['angle of attack (o)'] = np.where(df2['angle of attack (o)']>upper_limit,upper_limit,df2['angle of attack (o)'])
df2['angle of attack (o)'] = np.where(df2['angle of attack (o)']<lower_limit,lower_limit,df2['angle of attack (o)'])
print("After:", df2.shape)
sns.boxplot(df2['angle of attack (o)'],orient='v',ax=axes[1])
axes[1].title.set_text("After")
plt.show()

df = df2
```

* Since 'Chord Length' and 'Free Stream Velocity' are collected in the dataset as discrete values. We cannot expect those features to be fully normal distributions.
* Therefore, we will only need to transform following three features.

1. Suction Side Displacement Thickness - Right Skewed

3. Frequency - Right Skewed

<br/>

* Before applying data tranformations, let's split the data into training and testing.
* So that when we test our models, they will test on real-world data, not transformed values.
* The dataset is divided randomly into training and testing containing 80% and 20% instances.





```python
# Seperate dataset into training and testing

# Defining independent and dependent features
x = df.drop("scaled sound pressure level (db)", 1)
y = pd.DataFrame(df['scaled sound pressure level (db)'], columns=["scaled sound pressure level (db)"])

# Spliting dataset
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 101)

# Reset indexes
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

*   Applying logarithmic tranformations for 'Frequency', and 'Suction Side Displacement Thickness' columns. Since these two columns only have positive values, we can reduce the skewness with logarithmic transformation.

```python
# Define required columns for logarithm transformation
columns = ['frequency (Hz)', 'suction side displacement thickness (delta)']

# Initialize log transformer with logarithmic transformation 
log = FunctionTransformer(np.log, validate=True)

# Applying the transformation for train data
tansformed_df_train = log.transform(X_train[columns])

# Define new data frame
df_new_train = pd.DataFrame(tansformed_df_train, columns=columns)

# Replace X_train data with new values
X_train['frequency (Hz)'] = df_new_train['frequency (Hz)']
X_train['suction side displacement thickness (delta)'] = df_new_train['suction side displacement thickness (delta)']

# Applying the transformation for test data
tansformed_df_test = log.transform(X_test[columns])

# Define new data frame
df_new_test = pd.DataFrame(tansformed_df_test, columns=columns)

# Replace X_test data with new values
X_test['frequency (Hz)'] = df_new_test['frequency (Hz)']
X_test['suction side displacement thickness (delta)'] = df_new_test['suction side displacement thickness (delta)']
```

* Since Angle of Attack variable has zero values we can apply square root transformations instead of logarithmic transformation.

```python
# Define required columns for logarithm transformation
columns = ['angle of attack (o)']

# Initialize log transformer with logarithmic transformation 
sqrt_trans = FunctionTransformer(np.sqrt, validate=True)

# Applying the transformation for train data
tansformed_df_train = sqrt_trans.transform(X_train[columns])

# Define new data frame
df_new_train = pd.DataFrame(tansformed_df_train, columns=columns)

# Replace X_train data with new values
X_train['angle of attack (o)'] = df_new_train['angle of attack (o)']

# Applying the transformation for test data
tansformed_df_test = sqrt_trans.transform(X_test[columns])

# Define new data frame
df_new_test = pd.DataFrame(tansformed_df_test, columns=columns)

# Replace X_test data with new values
X_test['angle of attack (o)'] = df_new_test['angle of attack (o)']
```

## Dimensionality Reduction

* In this step, let's try to compress our dataset by removing redundant features.
* Let's apply PCA.



```python
pipe4 = Pipeline([('pca', PCA())])
param_grid = {
    'pca__n_components': [2,3,4,5],
    }

gcv = GridSearchCV(pipe4, param_grid, n_jobs=-1)
gcv.fit(X_train, y_train)
 
print('Best Parameter:', gcv.best_params_)
```

*  Since PCA has sorted the components according to covarianece, we can check covariance by reducing one component. Still 99% covariance can be achieved.
* Therefore we can define PCA with 4 components.



```python
pca = PCA(n_components = 4)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

## Polynomial Features

* Let's apply Polynomial Features to our dataset.
* After trial and error attempts degree=7 resulted in better accuracy.


```python
trans = PolynomialFeatures(degree=7)
X_train_pca = trans.fit_transform(X_train_pca)
X_test_pca = trans.fit_transform(X_test_pca)
```

## (c) Predict Y Using Regression Techniques

## I. Linear Regression With Cross Validation 

*   Let's apply Linear Regression with K Fold Cross Validation.



```python
linearModel = LinearRegression()  # create object for the class
model = linearModel.fit(X_train_pca, y_train)  # perform linear regression
y_pred_linear = linearModel.predict(X_test_pca)  # make predictions
y_hat_linear = pd.DataFrame(y_pred_linear, columns=["predicted"])
```

```python
X_combined = np.r_[X_train_pca, X_test_pca]
y_combined = np.r_[y_train, y_test]
scores = cross_val_score(model, X_combined, y_combined, cv=10)
print("Cross-validated scores:", scores)
```

```python
predictions = cross_val_predict(model, X_combined, y_combined, cv=10)
linear_accuracy = metrics.r2_score(y_combined, predictions)
print("Linear Model Cross-Predicted Accuracy : %.3f " % (linear_accuracy * 100) + "%")
```

```python
plt.figure(figsize=(15, 8))
plt.plot(y_hat_linear, label = "Pred")
plt.plot(y_test, label = "Actual")
plt.xlabel('x - axis')
# Set the y axis label of the current axis.
plt.ylabel('y - axis')
# Set a title of the current axes.
plt.title('Predicitions vs Actual')
# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```

## II. Lasso Regression With Cross Validation

```python
# List to maintain the cross-validation scores 
cross_val_scores_lasso = [] 
  
# List to maintain the different values of delta 
delta = [] 
  
# Loop to compute the cross-validation scores 
for i in range(0, 5): 
    lassoModel = Lasso(alpha = i *0.25) 
    lassoModel.fit(X_train_pca, y_train) 
    scores = cross_val_score(lassoModel, X_test_pca, y_test, cv = 10) 
    avg_cross_val_score = mean(scores)*100
    cross_val_scores_lasso.append(avg_cross_val_score) 
    delta.append(i * 0.25) 
  
# Loop to print the different values of cross-validation scores 
for i in range(0, len(delta)): 
    print(str(delta[i])+' : '+str(cross_val_scores_lasso[i]))
```

```python
# Building and fitting the Lasso Regression Model 
lassoModelChosen = Lasso(alpha = 0.0) 
lassoModelChosen.fit(X_train_pca, y_train)
y_pred_lasso = lassoModelChosen.predict(X_test_pca)  # make predictions
  
# Evaluating the Lasso Regression model 
print(lassoModelChosen.score(X_test_pca, y_test))
```

```python
scores = cross_val_score(lassoModelChosen, X_combined, y_combined, cv=10)
print("Cross-validated scores:", scores)
```

```python
predictions = cross_val_predict(lassoModelChosen, X_combined, y_combined, cv=10)
lasso_accuracy = metrics.r2_score(y_combined, predictions)
print("Lasso Model Cross-Predicted Accuracy : %.3f " % (lasso_accuracy * 100) + "%")
```

## III. Ridge Regression With Cross Validation

```python
# List to maintain the different cross-validation scores 
cross_val_scores_ridge = [] 
# List to maintain the different values of delta 
delta = [] 
# Loop to compute the different values of cross-validation scores 
for i in range(1, 10):
    ridgeModel = Ridge(alpha = i * 0.01) 
    ridgeModel.fit(X_train_pca, y_train)
    # Since K-Fold=10
    scores = cross_val_score(ridgeModel, X_test_pca, y_test, cv = 10)
    avg_cross_val_score = mean(scores)*100
    cross_val_scores_ridge.append(avg_cross_val_score) 
    delta.append(i * 0.01)
  
# Loop to print the different values of cross-validation scores 
for i in range(0, len(delta)): 
    print(str(delta[i])+' : '+str(cross_val_scores_ridge[i]))
```

```python
# Building and fitting the Ridge Regression model 
ridgeModelChosen = Ridge(alpha = 0.01) 
ridgeModelChosen.fit(X_train_pca, y_train) 
y_pred_ridge = ridgeModelChosen.predict(X_test_pca)  # make predictions

# Evaluating the Ridge Regression model 
print(ridgeModelChosen.score(X_test_pca, y_test)) 
```

```python
scores = cross_val_score(ridgeModelChosen, X_combined, y_combined, cv=10)
print("Cross-validated scores:", scores)
```

```python
predictions = cross_val_predict(ridgeModelChosen, X_combined, y_combined, cv=10)
ridge_accuracy = metrics.r2_score(y_combined, predictions)
print("Ridge Model Cross-Predicted Accuracy : %.3f " % (ridge_accuracy * 100) + "%")
```
